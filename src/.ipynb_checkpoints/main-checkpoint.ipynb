```python
import numpy as np
import pandas as pd
```

## 数据概况的观察
    - 使用pd.read_csv()完成数据的提取
    - 使用DataFrame.describe()完成数据概况的观察
## 数据列的提取
    - 把pd.read_csv()提取出来的DataFrame对象保存起来

```python
path1 = "./data/customer.csv"
path2 = "./data/holiday_mst.csv"
path3 = "./data/hotel.csv"
path4 = "./data/reserve.csv"

customer = pd.read_csv(path1)
holiday_mst = pd.read_csv(path2)
hotel = pd.read_csv(path3)
reserve = pd.read_csv(path4)

print('表customer:')
print(customer)
print(customer.describe())
print('\n\n\n')

print('表holiday_mst:')
print(holiday_mst)
print(holiday_mst.describe())
print('\n\n\n')

print('表hotel:')
print(hotel)
print(hotel.describe())
print('\n\n\n')

print('表reserve:')
print(reserve)
print(reserve.describe())
print('\n\n\n')
```

## 异常值的识别和剔除
    - 使用Z-score方法剔除异常值
## 规整化
    - 属性字段数值化,将man处理为1,woman处理为0,并且把bool值处理为0/1值
    - 归一化,把数据都归一化到[0,1]区间中,min值对应0,max值对应1
## 缺失值的填充
    - 使用聚合值填充,这里使用中位数
    
**注意:一定要在剔除异常值并且规整化之后后再填充缺失值**

```python
# 对表customer使用Z-Score方法
zscore = (customer - customer.mean()) / customer.std()
zscore_r = zscore[abs(zscore) < 3]

# 对表customer进行归一化处理
new_customer = zscore_r / (zscore_r.max() - zscore_r.min())
print('归一化后的new_customer:')
new_customer['customer_id'] = customer['customer_id']
new_customer['sex'] = customer['sex']
print(new_customer)
```

```python
# 表holiday_mst无需进行归一化处理
new_holiday_mst = holiday_mst

print(new_holiday_mst)
```

```python
# 对表hotel进行Z-Score方法
zscore = (hotel - hotel.mean()) / hotel.std()
zscore_r = zscore[abs(zscore) < 3]

# 对表hotel进行归一化处理
new_hotel = zscore_r / (zscore_r.max() - zscore_r.min())
print('归一化后的new_hotel:')
new_hotel['hotel_id'] = hotel['hotel_id']
new_hotel['big_area_name'] = hotel['big_area_name']
new_hotel['small_area_name'] = hotel['small_area_name']
print(new_hotel)
```

```python
# 对表reserve进行Z-Score方法
zscore = (reserve - reserve.mean()) / reserve.std()
zscore_r = zscore[abs(zscore) < 3]

# 对表reserve进行归一化处理
new_reserve = zscore_r / (zscore_r.max() - zscore_r.min())
print('归一化后的new_reserve:')
new_reserve['checkin_date'] = reserve['checkin_date']
new_reserve['checkin_time'] = reserve['checkin_time']
new_reserve['checkout_date'] = reserve['checkout_date']
new_reserve['customer_id'] = reserve['customer_id']
new_reserve['hotel_id'] = reserve['hotel_id']
new_reserve['reserve_datetime'] = reserve['reserve_datetime']
new_reserve['reserve_id'] = reserve['reserve_id']
print(new_reserve)
```

```python
# 表customer需要数据的规整化
customer = new_customer.replace('man',1)
customer = new_customer.replace('woman',0)

# 表holiday_mst需要数据的规整化
holiday_mst = new_holiday_mst.replace(True,1)
holiday_mst = new_holiday_mst.replace(False,0)

# 表hotel需要数据的规整化
hotel = new_hotel.replace(True,1)
hotel = new_hotel.replace(False,0)

# 表reserve不需要数据的规整化

print('表customer:')
print(customer)
print(customer.describe())
print('\n\n\n')

print('表holiday_mst:')
print(holiday_mst)
print(holiday_mst.describe())
print('\n\n\n')

print('表hotel:')
print(hotel)
print(hotel.describe())
print('\n\n\n')

print('表reserve:')
print(reserve)
print(reserve.describe())
print('\n\n\n')
```

```python
# 求这一行的聚合值,这里以中位数为聚合值,并用这些值填充None
customer_median = customer.median()
print(customer_median,end='\n\n')
customer = customer.fillna(customer_median)
print(customer)

holiday_mst_median = holiday_mst.median()
print(holiday_mst,end='\n\n')
holiday_mst = holiday_mst.fillna(holiday_mst_median)
print(holiday_mst)

hotel_median = hotel.median()
print(hotel_median,end='\n\n')
hotel = hotel.fillna(hotel_median)
print(hotel)

reserve_median = reserve.median()
print(reserve_median,end='\n\n')
reserve = reserve.fillna(reserve_median)
print(reserve)
```
